```python
from minio import Minio

client = Minio(
    "localhost:9000",  # Thay bằng IP mạng Docker nếu MinIO chạy ở container khác
    access_key="dataNLPmining-lab",
    secret_key="dataNLPmining-lab",
    secure=False
)
print ("connect success")
```

```python
from pyspark.sql import SparkSession

# Khởi tạo SparkSession với cấu hình kết nối MinIO (qua giao thức S3A)
spark = SparkSession.builder \
    .appName("PySpark-MinIO-Connection") \
    .config("spark.jars.packages", "org.apache.hadoop:hadoop-aws:3.3.4,com.amazonaws:aws-java-sdk-bundle:1.12.262") \
    .config("spark.hadoop.fs.s3a.endpoint", "http://127.0.0.1:9001/") \
    .config("spark.hadoop.fs.s3a.access.key", "admin") \
    .config("spark.hadoop.fs.s3a.secret.key", "password123") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false") \
    .getOrCreate()

print("SparkSession đã được khởi tạo thành công với cấu hình MinIO!")
```

```python
# Ví dụ đọc một file dữ liệu từ MinIO vào Spark DataFrame
df = spark.read.json("s3a://raw-financial-data/json-news/2026-05-30_fpt.json")
df.show()
```
